## 1. Unsupervised Learning

Unsupervised Learning uses Machine Learning algorithms to analyze and cluster unlabeled datasets and discover hidden patterns or data groupings without human intervention. There are two types of Unsupervised Learning algorithms: **Clustering** and **Association Rules**. K-means algorithm belongs to the clustering methods. 

## 2. Clustering

Clustering is one of the most common exploratory data analysis techniques used to understand the data structure. Clustering algorithms allow you to discover homogeneous subgroups within the dataset so that data points in each cluster are comparable and similar to each other based on a similarity measure like euclidean-based distance or correlation-based distance.

## 3. K-Means Clustering Algorithm

The K-means is an Unsupervised Machine Learning algorithm that splits a dataset into K non-overlapping subgroups (clusters). It allows us to split the data into different groups or categories. For example, if K=2 there will be two clusters, if K=3 there will be three clusters, etc. Using the K-means algorithm is a convenient way to discover the categories of groups in the unlabeled dataset.

The aim of the K-means algorithm (also known as a centroid-based algorithm) is to minimize the sum of distances between the data point and their corresponding clusters while associating each group of objects (cluster) with its centroid.

The K-means algorithm does two things:

- Iteratively determines the best value for centroids (center points for every Ks cluster)
- Assigns each data point to its closest K-center. The nearest to K-ceter data points form a cluster

As a result, each cluster will have data points with similar features:




Heres a high-level K-means algorithm description:

1. Define the number of clusters based on provided K value
2. Select random K points or centroids (can differ from the input dataset)
3. Form the K clusters by assigning each data point to their closest centroid
4. Calculate the variance and define a new centroids of each cluster
5. Repeat the process from the third step to reassign each datapoint to the new closest centroid of each cluster untill the algorithm finds best possible solution

## 4. How k-means clustering works?

**Step 1**: We start by selecting random points in the dataset. These will be the initial centroids (the central point of a cluster).




    
**Step 2**: Now we go over every point and assign it to a cluster. We want to assign the point to the nearest centroid. We do this by calculating the euclidean distance between that point and all centroids, then assigning the point to the nearest one.





**Step 3**: After all points have been assigned to a cluster, we recalculate the clusters centroid. We do this by calculating the group mean. This will shift the centroid to the center of the cluster:



We repeat step 2 and 3 until the centroids have converged (the centroids are the same as in the previous iteration) or a maximum number of iterations have been reached.
